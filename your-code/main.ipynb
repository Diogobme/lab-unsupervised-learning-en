# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Your code here:

# Supondo que j√° temos df_log (com log1p aplicado)
num_cols = Wholesale_customers_clean.select_dtypes(include=['float64', 'int64']).columns

# Inicializa o scaler
scaler = StandardScaler()

# Ajusta e transforma
customers_scale = scaler.fit_transform(Wholesale_customers_clean[num_cols])

# Transforma novamente em DataFrame (para visualiza√ß√£o)
customers_scale = pd.DataFrame(customers_scale, columns=num_cols)

print(customers_scale.head())
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Your code here:
# feature_cols = ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']

# 1) Escalar APENAS as 6 features (j√° transformadas com log1p em Wholesale_customers_clean)
X_log6 = Wholesale_customers_clean[feature_cols].copy()

scaler_k6 = StandardScaler()
customers_scale = scaler_k6.fit_transform(X_log6)   # shape: (n, 6)
assert customers_scale.shape[1] == len(feature_cols)

# 2) K-Means nas mesmas 6 colunas
kmeans = KMeans(n_clusters=3, n_init=20, random_state=42)
labels = kmeans.fit_predict(customers_scale)

# 3) Adicionar labels ao DF original (mesmo n√∫mero de linhas!)
Wholesale_customers['labels'] = labels
assert len(Wholesale_customers) == len(labels)

# 4) Centros -> desfaz escala -> desfaz log, usando o MESMO scaler_k6 (treinado em 6 cols)
centers_scaled = kmeans.cluster_centers_                 # (3, 6)
centers_log = scaler_k6.inverse_transform(centers_scaled)# (3, 6)
centers_orig = np.expm1(centers_log)                     # (3, 6)

# 5) Agora bate com feature_cols (6)
centers_df = pd.DataFrame(
    centers_orig,
    columns=feature_cols,
    index=[f"cluster_{i}" for i in range(3)]
)

print("Centros (escala original, arredondados):")
print(centers_df.round(0))

# Checks √∫teis para evitar o mesmo erro no futuro:
assert centers_df.shape == (3, len(feature_cols)), "Dimens√£o dos centros n√£o bate com feature_cols!"


```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2).fit(customers_scale)

labels = kmeans_2.predict(customers_scale)

clusters = kmeans_2.labels_.tolist()
```

```python
Wholesale_customers_clean['Label'] = clusters
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

# Your code here
# DBSCAN com eps=0.5 (padr√£o do enunciado) e min_samples=5 (default)
db = DBSCAN(eps=0.5, min_samples=5, metric='euclidean')

# Ajusta e obt√©m os r√≥tulos
db.fit(customers_scale)
labels_db = db.labels_   # -1 indica "ru√≠do" (outliers n√£o clusterizados)

# Anexa ao dataset original
Wholesale_customers['labels_DBSCAN'] = labels_db

# Resumo r√°pido
print("Contagem por label (inclui -1 como ru√≠do):")
print(Wholesale_customers['labels_DBSCAN'].value_counts().sort_index())

# Quantos clusters v√°lidos (excluindo ru√≠do)?
n_clusters = len(set(labels_db)) - (1 if -1 in labels_db else 0)
n_noise = np.sum(labels_db == -1)
print(f"\nClusters encontrados (excl. ru√≠do): {n_clusters}")
print(f"Pontos marcados como ru√≠do (-1): {n_noise}")
```

Count the values in `labels_DBSCAN`.

```python
# Your code here
# Count the values in labels_DBSCAN

print("Contagem de registros por cluster (DBSCAN):")
print(Wholesale_customers['labels_DBSCAN'].value_counts().sort_index())
```
